**Python set**

- `Python set concept is represented with {}`
- `Python set concept is an unordered pair `
- `Python set is not having Index concept | If there is no index concept | slicing and skipping concept is also not possible `
- `Python set concept will not allow duplicates `
- `Python set is Mutable But it will not allow mutables inside it `

```python
a = 'data' # save in the Memory
print(a) # calling from memory
```

**Python set Built_in_functions**

**Increase**
- `add`
- `update `

**Decrease**
- `pop `
- `remove `
- `discard`

**Some other Operations**
- `clear `
- `copy `
- `union `
- `intersection `

```python
t = [10,20,30]

t.append()

print(t)
```

```python
# add = will take only one input from user
# and add to Memory randomlly


f = {10,50,9,19} # save in the Memory

f.add(500)

print(f) # calling from Memory
```

```python
# update = will take multiple values from user
# and add randomlly

f = {10,50,9,19} # save in the Memory

f.update([100,200,500])

print(f) # calling from Memory
```

```python
# pop = will remove one value from existing set randomlly

f = {10,50,9,19} # save in the Memory

f.pop()

print(f) #calling from Memory
```

```python
# discard = this function will take a value from user
# and if the value is present in existing set it will remove
# or if value is not present it will be silent


f = {10,50,9,19} # save in the Memory

f.discard(50)

print(f) # calling from Memory
```

```python
# Remove = this function will take a value from user
# and if the value is present in existing set it will remove
# or if value is not present it wont be silent


f = {10,50,9,19} # save in the Memory

f.remove(50)

print(f) # calling from Memory
```

```python
# discard = this function will take a value from user
# and if the value is present in existing set it will remove
# or if value is not present it will be silent


f = {10,50,9,19} # save in the Memory

f.discard(550)

print(f) # calling from Memory
```

**From the Above Operations we can say that Python Set concept is Mutable**

```python
a = {1,2,3,4,5,6}
b = {5,6,7,8,9,10}

print(a.union(b))
```

```python
print(a.intersection(b))
```

```python
# copy

a = {10,50,90,100}
b = a # deep copy

print(a)
print(b)

print(id(a))
print(id(b))

a.add(560)

print(a)
print(b)
```

```python
# copy

a = {10,50,90,100}
b = a.copy() # shallow copy

print(a)
print(b)

print(id(a))
print(id(b))

a.add(560)

print(a)
print(b)
```

```python
a = [10,20,30,50] # save in the Memory

a.clear()

print(a) # calling from Memory
```

```python
a = {10,20,30,50} # save in the Memory

a.clear()

print(a) # calling from Memory
```
