**Python Dictionary**

- `Python dictionary is represented with -> {}`
- `Anything we write inside the {} -> it becomes set but inside {} -> if we write key_value pair then it is a dictionary`
- `Key means column name and value means data inside the column `
- `we can create the exsheet format data using dictionary but we cant see -> if we give the data to pandas library we can see the elsheet `
- `we can pass anything in the place of values `
- `we can pass anything in place of keys except the mutable types `
- `there will be no index concept in dict if there is no index there will be no slicing and no skipping `

**Python Dictionary concept is Mutable or Immutable -> prove it**

```python
# try to increase the Memory size

fruits = {'apple':100 , 'banana': 200 ,'carrot':300}

fruits['apple'] = 500

print(fruits)
```

```python
# try to increase the Memory size

fruits = {'apple':100 , 'banana': 200 ,'carrot':300}

fruits['grapes'] = 500

print(fruits)
```

```python
# try to increase the Memory size

fruits = {'apple':100 , 'banana': 200 ,'carrot':300}

del fruits['banana']

print(fruits)
```

```python
a = {'Name' : ['John','Peter','Steve'] ,
     'Age' : [27,45,45] ,
     'Name' : [1200.85 , 1500.63 , 1900.22]
     } # using dict we can design the data in the form of exsheet
print(a)
```

```python
b = a['Age']
print(b.count(45))
```

**From the Above Operations we can say that Python Dict is Mutable**

```python
# get

a = {'Name':'John' , 'Age':27 , 'Salary':1200.85} # save in the Memory

print(a['Age'])
```

```python
# get

a = {'Name':'John' , 'Age':27 , 'Salary':1200.85} # save in the Memory

print(a.get('Age'))
```

```python
a = {'Name' : ['John','Peter','Steve'] ,
     'Age' : [27,45,55] ,
     'salary' : [1200.85 , 1500.63 , 1900.22]
     } # using dict we can design the data in the form of exsheet

print(a['salary'])
```

```python
a = {'Name' : ['John','Peter','Steve'] ,
     'Age' : [27,45,55] ,
     'salary' : [1200.85 , 1500.63 , 1900.22]
     } # using dict we can design the data in the form of exsheet

print(a.get('salary'))
```

```python
# update:
fruits = {'apple':100 , 'banana': 200 ,'carrot':300}

fruits['grapes'] = 500

print(fruits)

```

```python
fruits = {'apple':100 , 'banana': 200 ,'carrot':300}

fruits.update({'grapes':500})

print(fruits)
```

```python
# pop

fruits = {'apple':100 , 'banana': 200 ,'carrot':300}

del fruits['banana']

print(fruits)
```

```python
# pop

fruits = {'apple':100 , 'banana': 200 ,'carrot':300}

fruits.pop('banana')

print(fruits)
```

```python
# popitem : remove last key

fruits = {'apple':100 , 'banana': 200 ,'carrot':300}

fruits.popitem()

print(fruits)
```

```python
fruits = {'apple':100 , 'banana': 200 ,'carrot':300}

print(fruits.keys())
```

```python
fruits = {'apple':100 , 'banana': 200 ,'carrot':300}

print(fruits.values())
```

```python
fruits = {'apple':100 , 'banana': 200 ,'carrot':300}

print(fruits.items())
```

```python
fruits = {'apple':100 , 'banana': 200 ,'carrot':300}

fruits.clear()


print(fruits)
```

```python
fruits = {'apple':100 , 'banana': 200 ,'carrot':300}
b = fruits # deep copy

print(fruits)
print(b)

print(id(fruits))
print(id(b))

fruits['apple'] = 500

print(fruits)
print(b)
```

```python
fruits = {'apple':100 , 'banana': 200 ,'carrot':300}
b = fruits.copy() # shallow copy

print(fruits)
print(b)

print(id(fruits))
print(id(b))

fruits['apple'] = 500

print(fruits)
print(b)
```
