**Python List Built_in_functions**

**additon**
- `append `
- `extend `
- `insert`

**deletion**
- `pop `
- `remove `

**Some common operations**
- `index `
- `count `
- `copy `
- `clear `
- `sort `
- `reverse `

```python
# append = append is a built_in_function which is used to take
# one single input from user and add to the end of the existing List

a = [10,50,12,90] # save in the Memory

a.append(500)

print(a) # calling from memory
```

```python
# extend = this function will take multiple inputs from the user
# and add to the end of the existing List


a = [10,50,12,90] # save in the Memory

a.extend( [100,500,600] )

print(a) # calling from Memory
```

```python
# insert = we can add the value where we want

a = [10,50,12,90] # save in the Memory

a.insert(1,100)
a.insert(3,200)

print(a) # calling
```

```python
# pop = pop defautly will remove the last value
# but if we pass an input it will consider as index and will be removed

a = [10,50,12,90] # save in the Memory

a.pop()

print(a) # calling from Memory
```

```python
# pop = pop defautly will remove the last value
# but if we pass an input it will consider as index and will be removed

a = [10,50,12,90] # save in the Memory

a.pop(2)

print(a) # calling from Memory
```

```python
# remove = pass a value and remove it

a = [10,50,12,90] # save in the Memory


a.remove(12)

print(a) # calling
```

**from the above operations we can say that Python List is Mutable**

```python
# count

a = [10,50,12,90,50,10] # save in the Memory

print(a.count(50))
```

```python
# index

a = [10,50,12,90,50,10] # save in the Memory

print(a.index(50,2))
```

```python
# reverse -> [::-1]

a = [10,20,50] # save in the Memory

a.reverse()

print(a) # calling from Memory
```

```python
# sort

a = [10,5,8,12,4,6] # save in the Memory

a.sort() # ascending order

print(a) # calling from Memory
```

```python
# sort

a = [10,5,8,12,4,6] # save in the Memory

a.sort(reverse=True) # descending order

print(a) # calling from Memory
```

```python
a = [10,50,100] # save in the Memory

del a

print(a) # calling from memory
```

```python
# clear = this function wil remove all the values from the varibale
# but varibale will be alive

a = [10,50,100] # save in the Memory

a.clear()

print(a) # calling from Memory
```

```python
# copy

a = [10,50,100] # save in the Memory
b = a # deep copy

print(a)
print(b)

print(id(a))
print(id(b))

a.append(1000)

print(a)
print(b)
```

```python
# copy

a = [10,50,100] # save in the Memory
b = a.copy() # shallow copy

print(a)
print(b)

print(id(a))
print(id(b))

a.append(1000)

print(a)
print(b)
```
