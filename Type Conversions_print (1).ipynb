**Type Conversions -> Data type Conversions**

- `int -> float and int -> str`
- `float -> int and float -> str `
- `str -> int and str -> float `

```python
a = '156'
b = float(a)
b
```

```python
# int -> float

a = 22 # save in the Memory
print(a) # calling from Memory
print(type(a)) #calling and checking the data type

b = float(a) # actual conversions happends here
print(b) # calling from memory
print(type(b)) # calling and chekcing the data type
```

```python
a = 22
b = str(a)
print(b)
```

**print Conditions**

`to create the relation with <str and int> | <str and float> inside the print condition `

`step:1 -> Using comma (,) where , will give us one tab space => only 10% developers will use`

`Step:2 -> Using (+) -> it will not give one tab space and + will create the relation with str only  `

`step: 3 -> using identifiers %d-> int %f float %s string `

`step: 4 -> using f and {}`

```python
# identifiers

f = 27 # save in the Memory

print('My name is Ajay and My age is %d'%(f))
```

```python
course = 'data science'
time = 6
duration = 1.30

# expected output: I am learning data science study in vihara tech for 6 months
                   # with 1.5 hours duration daily
print('I am learning %s study in VT for %d months with %f hours duration daily'%(duration,time,course))
```
