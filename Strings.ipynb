**what is Mutable and Immutable**

`once a data types saves in Memory if developer increases the Memory size then it is a Mutable and if we cant increase the Memory size then it is Immutable `

**String in Python is Mutable or Immutable -> Prove it**

- `try to increase the Existing Memory size`
- `try to decrease the Existing Memory size `

```python
# try to increase the Memory size

f = 'data' # save in the Memory

f[4] = 'j'

print(f) # calling from Memory
```

```python
# try to decrease the Memory size

f = 'data' # save in the Memory

del f[2]

print(f) # calling from Memory
```

**From the Above Operations we can say that Python strings are Immutable**

```python
# capitalize = capitalize built_in_function will convert entire paragragh into small
               # and only starting character into capital
t = 'inDIA CHIna' # save in the Memory

t = t.capitalize()

print(t) # calling from Memory
```

```python
# title = will convert each word first letter into capital
# remaining all are small

t = 'inDIA cHIna' # save in the Memory
t = t.title()
print(t)
```

```python
# lower = lower function is used to convert into small letter
f = 'daTA SCIEnce' # save in the Memory
f = f.lower()
print(f) # calling from Memory
```

```python
# upper = upper function is used to convert into capital letter
f = 'daTA SCIEnce' # save in the Memory
f = f.upper() # in this line the Memory will be updated
print(f) # calling from memory
```

```python
# isupper = will return the boolean outcome

f = 'daTA SCIENce'

print(f.isupper())
```

```python
# islower = will return the boolen outcome

f = 'data science'

print(f.islower())
```
